```python
# Load resources
import pandas as pd
```

```python
# Load file
statsFile = "../resources/international_airline_activity_monthly_airline_performance_1120.xlsx"
ittfa_df = pd.read_excel(statsFile)
ittfa_df.head(10)
```

```python
# Rename Columns

ittfa_df.columns = ['year','month','passengers_in','passengers_out','passengers_total','drop_1','freight_in','freight_out','freight_total','drop_2','mail_in','mail_out','mail_total','drop_3','aircraft_in','aircraft_out','aircraft_total','drop_4','flights_in','flights_out','flights_total','drop_5','passengers_incl_transit_in','passengers_incl_transit_out','passengers_incl_transit_total','drop_6','available_seats_in','available_seats_out','available_seats_total','drop_7','seat_utilisation_in','seat_utilisation_out','seat_utilisation_total']
ittfa_df.head(10)
```

```python
# Drop irrelevant columns
dropCols = [col for col in ittfa_df.columns if 'drop' in col]
dropCols

clean_ittfa_df = ittfa_df.drop(dropCols,axis=1)
clean_ittfa_df.head(20)
```

```python
clean_ittfa_df['year'] = clean_ittfa_df[clean_ittfa_df['year'].apply(lambda x: str(x).isdigit())]
clean_ittfa_df.head(20)
```

```python
clean_ittfa_df['year'] = pd.to_numeric(clean_ittfa_df['year'])
clean_ittfa_df.dtypes
clean_ittfa_df = clean_ittfa_df.dropna()
```

```python
# Drop all rows year < 2009

clean_ittfa_df = clean_ittfa_df.drop(clean_ittfa_df[clean_ittfa_df['year'] < 2009].index)
clean_ittfa_df['year'] = clean_ittfa_df['year'].astype(int)
clean_ittfa_df.head(50)
```

```python
# print(clean_ittfa_df.columns[2:])
for col in clean_ittfa_df.columns[2:]:
    clean_ittfa_df[col] = pd.to_numeric(clean_ittfa_df[col])

clean_ittfa_df.dtypes
```

```python
clean_ittfa_df.to_csv('../Resources/ittfa_2009-2020.csv',index=False)
```
